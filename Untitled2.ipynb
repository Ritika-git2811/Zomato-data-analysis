**Zomato Data Analysis Project**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('/Zomato.csv')
```

```python
print(df)
```

Convert the data type of column -**rate**


```python
df['rate']=df['rate'].str.split('/').str[0].astype(float)
```

```python
print(df)
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('type of resturants')
```

**Conclusion-majority of the resturant falls in dinning category **

```python
grouped_data= df.groupby('listed_in(type)') ['votes'].sum()
result =pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel('type of resturants',c='red', size=10)
plt.ylabel('votes', c='red', size=10)
```

**Conclusion- dinning resturants has max votes**

```python
plt.hist(df['rate'],bins=10)
plt.title('Rate Distribution')
plt.show()
```

**Conclusion-majority resturants has 3.5 to 4**

**Average order spended by couples**

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

**Conclusion -majority of couples prefer 300 cost**

**Which mode receives max rating**

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data= df)
```

Conclusion- offline order received lower rating in comparison of online rating

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed in (type)')
plt.show()
```

**Conclusion-Dining resturants primarily accept offline orders, where as cafes primarily receive online orders. This suggests that clients prefer online orders inperson at resturants,but prefer online ordering at cafes.**
